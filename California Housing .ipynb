```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(california_housing.data, california_housing.target)
```

```python
test_pred = lin_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
lin_mse = mean_squared_error(y_test, test_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_regression = DecisionTreeRegressor()
```

```python
tree_regression.fit(california_housing.data , california_housing.target)
```

```python
test_tree_pred = tree_regression.predict(x_test)
```

```python
lin_tree_mse = mean_squared_error(y_test, test_tree_pred)
lin_tree_rmse = np.sqrt(lin_tree_mse)
int(lin_tree_rmse)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score_cross_val_tree = cross_val_score(tree_regression,california_housing.data , california_housing.target, scoring = "neg_mean_squared_error", cv=10)
```

```python
tree_rmse_score = np.sqrt(-score_cross_val_tree)
tree_rmse_score
```

```python
score_cross_val_linear = cross_val_score(lin_reg,california_housing.data , california_housing.target, scoring = "neg_mean_squared_error", cv=10)
```

```python
linear_rmse_score = np.sqrt(-score_cross_val_linear)
linear_rmse_score
```
